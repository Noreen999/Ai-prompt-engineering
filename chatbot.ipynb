```python
from transformers import DistilBertTokenizer, DistilBertModel
import torch
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-cased-distilled-squad')
model = DistilBertModel.from_pretrained('distilbert-base-cased-distilled-squad')

question, text = "Who was Jim Henson?", "Jim Henson was a nice puppet"

inputs = tokenizer(question, text, return_tensors="pt")
with torch.no_grad():
    outputs = model(**inputs)

print(outputs)
```

```python
from transformers import DistilBertTokenizer, TFDistilBertForQuestionAnswering
import tensorflow as tf

tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-cased-distilled-squad")
model = TFDistilBertForQuestionAnswering.from_pretrained("distilbert-base-cased-distilled-squad")

question, text = "Who was Jim Henson?", "Jim Henson was a nice puppet"

inputs = tokenizer(question, text, return_tensors="tf")
outputs = model(**inputs)

answer_start_index = int(tf.math.argmax(outputs.start_logits, axis=-1)[0])
answer_end_index = int(tf.math.argmax(outputs.end_logits, axis=-1)[0])

predict_answer_tokens = inputs.input_ids[0, answer_start_index : answer_end_index + 1]
tokenizer.decode(predict_answer_tokens)
```

```python
from transformers import pipeline

# Load QA pipeline
qa_pipeline = pipeline("question-answering")

# Predefined knowledge base for specific travel questions
travel_knowledge = {
    "what should i pack": "You should pack clothes, toiletries, travel documents, and a first-aid kit. Don't forget medications and chargers.",
    "visa requirements for france": "For France, you may need a Schengen visa depending on your nationality.",
    "best destinations in asia": "Some of the best destinations in Asia are Bali, Tokyo, Bangkok, and Kyoto.",
    "how to stay safe while traveling": "Always stay aware of your surroundings, keep your valuables secure, and follow local guidelines. Travel insurance is also recommended."
}

# Context for open-ended travel-related questions
context = """
Traveling requires planning and preparation. Always pack essentials like clothes, toiletries, and travel documents.
Ensure you have travel insurance, check the visa requirements for your destination, and stay updated on safety guidelines.
Top destinations include Paris, New York, Tokyo, and Bali. Always follow local customs and regulations while traveling.
"""

# Chatbot introduction
print("Chatbot: Hi! Ask me anything about traveling. Type 'exit' to end the chat.")

while True:
    # Get user input
    user_input = input("You: ").strip().lower()

    # Exit condition
    if user_input in ["exit", "quit", "bye"]:
        print("Chatbot: Goodbye! Have a safe journey!")
        break

    # Check predefined knowledge base for answers
    if user_input in travel_knowledge:
        print(f"Chatbot: {travel_knowledge[user_input]}")
    else:
        # Use QA pipeline for open-ended questions
        try:
            response = qa_pipeline(question=user_input, context=context)
            print(f"Chatbot: {response['answer']}")
        except Exception as e:
            print("Chatbot: Sorry, I couldn't understand your question. Could you rephrase it?")
```
